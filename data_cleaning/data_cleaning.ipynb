```python
import pandas as pd
```

```python

csv_path1 = "data/input/akc-data-latest.csv"
akc_df = pd.read_csv(csv_path1)
akc_df.head()
```

```python

csv_path2 = "data/input/kaggle_breeds_unique.csv"
kaggle_df = pd.read_csv(csv_path2)
kaggle_df.head()
```

```python
# breed replace spaces with _

akc_df['akc_breed'] = akc_df['breed']

akc_df['akc_breed'] = akc_df['akc_breed'].str.replace(' ', '_', regex=False)
akc_df.head()

kaggle_df['akc_breed'] = kaggle_df['kbreed']

kaggle_df['akc_breed'] = kaggle_df['akc_breed'].str.replace(' ', '_', regex=False)
kaggle_df.head()
```

```python
# breed all lowercase
akc_df['akc_breed'] = akc_df['akc_breed'].str.lower()
akc_df.head()

kaggle_df['akc_breed'] = kaggle_df['akc_breed'].str.lower()
kaggle_df.head()
```

```python
# outer join to see all akc with matching kaggle

joined_data = akc_df.merge(kaggle_df, on='akc_breed', how='outer')
joined_data.head()
```

```python
# export to csv for manual review of non-matches
joined_data.to_csv ('data/output/joined_breeds.csv', encoding='utf-8', index=False)
```
