## Read in dataset to obtain the list of breeds

```python
#import pandas to read in csv 
import pandas as pd
```

```python
file_path = "data/input/akc-data-latest.csv"
df_dog_breeds = pd.read_csv(file_path, encoding='utf-8')

df_dog_breeds.head(10)
```

```python
# list of problematic dog names: [79, 122, 160, 189, 238, 239, 249, 261]
```

```python
# replace o umlaut with reg o
df_dogbreeds = df_dogbreeds.replace('Löwchen','Lowchen')
df_dogbreeds.iloc[160,0]
```

```python
# change Vendéen to Vendeen
df_dogbreeds = df_dogbreeds.replace('Grand Basset Griffon Vendéen', 'Grand Basset Griffon Vendeen')
df_dogbreeds.iloc[122,0]
```

```python
# change dell'Etna to delletna
df_dogbreeds = df_dogbreeds.replace('Cirneco dell’Etna', 'Cirneco delletna')
df_dogbreeds.iloc[79,0]
```

```python
# change Vendéen to Vendeen
df_dogbreeds = df_dogbreeds.replace('Petit Basset Griffon Vendéen', 'Petit Basset Griffon Vendeen')
df_dogbreeds.iloc[189,0]
```

```python
# change breed to be compatible with akc website
df_dogbreeds = df_dogbreeds.replace('Small Munsterlander Pointer', 'Small Munsterlander')
df_dogbreeds.iloc[239,0]
```

```python
#check datatypes of cols
df_dogbreeds.dtypes
```

```python
#rename unnamed: 0 column 
df_dogbreeds = df_dogbreeds.rename(columns={'Unnamed: 0': 'breeds'})
df_dogbreeds
```

```python
#convert datatype of breeds col to string
df_dogbreeds['breeds'] = df_dogbreeds.breeds.astype('string')
df_dogbreeds.dtypes
```

```python
# get breeds into a list to be used for webscraping
dog_breeds = df_dogbreeds['breeds']
dog_breeds
```

```python
# make variables compatible with website url params (replace space with hyphen)
breed_list =[]
for dog in dog_breeds:
    dog = dog.replace(' ','-')
    breed_list.append(dog)
print(len(breed_list))
```

```python
breed_list[:10]
```

```python
# create an empty list that will be used for df
db_img = []
```

```python
# set up loop for all dogs in dataset
for doggo in breed_list:
    # create specific url
    concat = url + doggo 
    
    # visit webpage
    browser.visit(concat)
    html = browser.html
    img_soup = soup(html, 'html.parser')
    
    # go through html txt to find wrap where img lives
    images = img_soup.find_all('li', class_='slick-active')
    
    # append image to a list
    for im in images:
        if im.img:
            if im.img.has_attr('data-src'):
                img_updated = im.img['data-src'].replace("'","")
            
    # make a dict
    doggies = {
        'Breed': doggo,
        'Img_Link': img_updated
    }
    # append dict to df list 
    db_img.append(doggies)
    # loop to next dog
```

```python
# check to see if loop worked
db_img[:10]
```

```python
# exit browser
browser.quit()
```

```python
# create dataframe
breed_img_df = pd.DataFrame(db_img)
```

```python
# find which breeds dont have correct imgs
dup = breed_img_df.duplicated(subset=['Img_Link'])
i = 0;
dup_list = []
for n in dup:
    if(n):
        dup_list.append(i)
    i = i + 1
print(dup_list)
```

```python
need_doggies = breed_img_df.iloc[dup_list,0]
need_dog_img = []
for dog in need_doggies:
    need_dog_img.append(dog)
need_dog_img
```

```python
# replace [238, 249, 261] images with their real one. Fake ones:
# https://www.akc.org/wp-content/uploads/2017/11/Slovensky-Cuvac-standing-outdoors.jpg
# https://www.akc.org/wp-content/uploads/2017/11/Sussex-Spaniel-On-White-01.jpg
# https://www.akc.org/wp-content/uploads/2017/11/Transylvanian-Hound-On-White-01.jpg
```

```python
#open browser
browser = Browser('chrome')
url = 'https://www.akc.org/dog-breeds/'

# make list for images for the right breed
right_breed = []

# go through loop to webscrape 
for pup in need_dog_img:
    concat = url + pup
    browser.visit(concat)

    html = browser.html
    img_soup = soup(html, 'html.parser')
    
    #slick-slide slick-active slick-current
    images = img_soup.find_all('div', class_='slick-slide slick-active slick-current')
    img_list_dup = []    
    # append image to a list
    for im in images:
        if im.img:
            if im.img.has_attr('data-src'):
                img_updated = im.img['data-src'].replace("'","")
                img_list_dup.append(img_updated)
    new_img = img_list_dup[0]
    # add img to list
    right_breed.append(new_img)
#exit browser
browser.quit()
```

```python
#check to see if loop worked
right_breed
```

```python
# replace fake img links
x = 0
for n in right_breed:
    breed_img_df.loc[dup_list[x], 'Img_Link'] = right_breed[x]
    x = x + 1
```
